```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.style as style
```

```python
import os
```

```python
def concat_csv():
    dfs = []
    current_path = os.getcwd()
    source_path = os.path.join(current_path, 'cleaned_csv')
    os.chdir(source_path)
    ls = os.listdir()
    for i in ls:
        if i.endswith('.csv'):
            df = pd.read_csv(i, usecols=[1,2,3,4,5,6,7])
            dfs.append(df)
    os.chdir(current_path)
    return dfs
```

```python
dfs = concat_csv()
```

```python
dfs
```

```python
data = pd.concat(dfs)
data
```
