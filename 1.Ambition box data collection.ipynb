```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def get_company_details(location):
    header = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36 Edg/123.0.0.0'}

    company_names = []
    company_ratings = []
    other_data = []
    
    for page_no in range(1,501):
        url =  'https://www.ambitionbox.com/list-of-companies?'
        
        params = {
            'sortBy':'popular',
            'locations':location,
            'page':page_no
        }
        
        resp = requests.get(url, params=params, headers=header)
        soup = BeautifulSoup(resp.text, 'html')
        
        all_divs = soup.find_all('div', attrs={'class':"companyCardWrapper__metaInformation"})
        
        for i in all_divs:
            if i.h2:
                company_names.append(i.h2.text.strip())
            else:
                company_names.append(None)
            if i.find('span', attrs={'class':"companyCardWrapper__companyRatingValue"}):
                company_ratings.append(i.find('span', attrs={'class':"companyCardWrapper__companyRatingValue"}).text)
            else:
                company_ratings.append(None)
            if i.find('span', attrs={'class':"companyCardWrapper__interLinking"}):
                other_data.append(','.join(i.find('span', attrs={'class':"companyCardWrapper__interLinking"}).text.strip().split('|')))
            else:
                other_data.append(None)
        print(f"{location} - {page_no}")
    else:
        df = pd.DataFrame({
            'company_name':company_names,
            'company_rating':company_ratings,
            'other_data':other_data
        })
        
        df.to_csv(f"./csv_files\{location}.csv")
```
