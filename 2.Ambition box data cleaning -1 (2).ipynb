```python
def filter_data(source, destination, current_path):
    os.chdir(source)
    ls = os.listdir()
    for i in ls:
        if i.endswith('.csv'):
            data = pd.read_csv(i)
            data.drop('Unnamed: 0', axis=1, inplace=True)
            data[data['other_data'].apply(lambda x: True if len(x.split(' , ')) == 5 else False)].to_csv(os.path.join(destination_path, i))
    os.chdir(current_path)
```

```python
import os
```

```python
current_path = os.getcwd()
source_path = os.path.join(os.getcwd(), 'final_csv')
destination_path = os.path.join(os.getcwd(), 'cleaned_csv')
```

```python
def add_columns(source, destination, current_path):
    os.chdir(source)
    ls = os.listdir()
    for i in ls:
        if i.endswith('.csv'):
            data = pd.read_csv(i, usecols=[1,2,3])
            df = data.assign(
                industry = data['other_data'].apply(lambda x: x.split(' , ')[0]),
                size = data['other_data'].apply(lambda x: x.split(' , ')[1]),
                type = data['other_data'].apply(lambda x: x.split(' , ')[2]),
                years_old = data['other_data'].apply(lambda x: x.split(' , ')[3]).apply(lambda s: s.split()[0]).astype('int'),
                location = data['other_data'].apply(lambda x: x.split(' , ')[4]).apply(lambda s: s.split()[0]),
                )
            df.drop('other_data', axis=1, inplace=True)
            df.to_csv(os.path.join(destination_path, i))
    os.chdir(current_path)
```

```python
add_columns(source_path, destination_path, current_path)
```
